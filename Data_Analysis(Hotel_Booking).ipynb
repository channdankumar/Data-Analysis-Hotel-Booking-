# Data Analysis and Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc
```

```python
import matplotlib.pyplot as plt

# we are  already imported pandas and loaded your data into the DataFrame 'df'

# Calculate the percentage of canceled and not canceled reservations
cancelled_perc = df['is_canceled'].value_counts(normalize=True) * 100

# Print the percentage
print(cancelled_perc)

# Plot the bar chart
plt.figure(figsize=(5, 5))
plt.title('Reservation status count')
plt.bar(['not canceled', 'canceled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

# Create a countplot with 'hotel' on the x-axis, 'is_canceled' on the hue, and using the 'df' DataFrame
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Get the legend handles and labels
legend_labels, _ = ax1.get_legend_handles_labels()

# Set the legend position to be outside the plot (on the top-right)
ax1.legend(bbox_to_anchor=(1, 1))

plt.title('Reservation status in different hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# we are  have already imported pandas and loaded your data into the DataFrame 'df'

# Convert 'reservation_status_date' column to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Extract the month from the 'reservation_status_date' and create a new 'month' column
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16, 8))

# Create a countplot with 'month' on the x-axis, 'is_canceled' on the hue, and using the 'df' DataFrame
ax1 = sns.countplot(x='month', hue='is_canceled', data=df)

plt.ylabel('Number of reservations')
plt.legend(['Not Canceled', 'Canceled'])

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already imported pandas and loaded your data into the DataFrame 'df'

# Convert 'reservation_status_date' column to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Extract the month from the 'reservation_status_date' and create a new 'month' column
df['month'] = df['reservation_status_date'].dt.month

# Create a DataFrame containing the sum of 'adr' per month for canceled reservations
canceled_adr_per_month = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(10, 5))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=canceled_adr_per_month)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data into the DataFrame 'df'
# For example, if you have a CSV file named 'data.csv':
# df = pd.read_csv('data.csv')

# Assuming you have loaded the data into the DataFrame 'df'

# Filter the DataFrame to select only canceled reservations
cancelled_data = df[df['is_canceled'] == 1]

# Count the number of canceled reservations for each country and select the top 10 countries
top_10_country = cancelled_data['country'].value_counts()[:10]

plt.figure(figsize=(8, 8))
plt.title('Top 10 countries with reservations canceled')
plt.pie(top_10_country, autopct='%.2f%%', labels=top_10_country.index)

plt.show()
```
